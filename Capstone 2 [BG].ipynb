# Step 1: Data Cleansing
- Import Python libraries and import 3 different CSV files for the 3 sectors that the U.S. Energy Information Administration provides Renewable Energy Consumption Data on: Industrial Sector, Residential and Commercial Sectors, and Transportation and Electric Power Sectors.
    - Each database was cleaned individually before being concatenated.
- 'MSN' column acronym's were compared to the corresponding description in the 'Description' column and renamed to reflect the renewable energy type - making 'MSN' more understandable and allowing easier comparison across the three databases.
- Turned the 'YYYYMM' column to two columns - 'Year' and 'Month' and turned the string into date objects. The databases had a month labeled '13' in the 'YYYYMM' column to represent yearly summaries. All '13' variables were turned into a numerical null. 
- With 'MSN' and dates adjusted, the databases were concatenated. The databases had the 'Value' column as a string because missing values were marked as 'Not Available', 'No Data Reported', or 'Not Meaningful'. These strings were turned into null values and the column objects were converted into floats.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
%matplotlib inline
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
df_industrial = pd.read_csv("Renewable Energy Consumption - Industrial Sector.csv")
df_residentcommercial = pd.read_csv("Renewable Energy Consumption - Residential and Commercial Sectors.csv")
df_transportelectric = pd.read_csv("Renewable Energy Consumption - Transportation and Electric Power Sectors.csv")
```

# Making Industrial data more readable:

```python
print(df_industrial.columns)

energy_sectors= df_industrial['MSN'].unique()
print(energy_sectors)

descriptions= df_industrial['Description'].unique()
print(descriptions)
```

```python
energy_sector_rename = {'HVICBUS': 'Hydroelectric Power', 'GEICBUS': 'Geothermal Energy', 'SOICBUS':'Solar Energy',
                        'WYICBUS':'Wind Energy', 'WDICBUS':'Wood Energy', 'WSICBUS':'Waste Energy', 'EMICBUS': 'Fuel Ethanol',
                        'BFLCBUS':'Biomass Losses and Co-products', 'BMICBUS':'Biomass Energy', 'REICBUS':'Total Renewable Energy'}
df_industrial['MSN'] = df_industrial['MSN'].replace(energy_sector_rename)
print(df_industrial['MSN'])
print(df_industrial)
```

```python
df_industrial['YYYYMM'] = df_industrial['YYYYMM'].astype(str)

df_industrial['Year'] = pd.to_datetime(df_industrial['YYYYMM'].str[:4], format='%Y')
df_industrial['Month'] = pd.to_numeric(df_industrial['YYYYMM'].str[4:])
df_industrial['Month'] = df_industrial['Month'].replace(13, np.nan)
df_industrial['Month'] = df_industrial['Month'].astype('Int64')
df_industrial['Month'] = df_industrial.apply(lambda row: pd.Timestamp(year=row['Year'].year, month=row['Month'], day=1) if pd.notna(row['Month']) else pd.NaT, axis=1)

print(df_industrial)
```

# Merging Industrial, Residential and Commercial, and Transportation and Electric Power Sector dataframes into one:

```python
renewable_energy_df = pd.concat([df_industrial,df_residentcommercial,df_transportelectric], ignore_index=True)
print(renewable_energy_df)
```

```python
renewable_energy_df['Value'] = renewable_energy_df['Value'].replace('Not Available', np.nan)
renewable_energy_df['Value'] = renewable_energy_df['Value'].replace('No Data Reported', np.nan)
renewable_energy_df['Value'] = renewable_energy_df['Value'].replace('Not Meaningful', np.nan)
renewable_energy_df['Value'] = renewable_energy_df['Value'].astype(float)
renewable_energy_df.head()
```

```python
## Now, let's double check that the MSN column doesn't have redundancies in energy names - which were given per the
## description of each row of data.
energy_types= renewable_energy_df['MSN'].unique()
print(energy_types)

renewable_energy_df['MSN'] = renewable_energy_df['MSN'].replace("Fuel Ethanol, Excluding Denaturant", "Fuel Ethanol")
energy_types= renewable_energy_df['MSN'].unique()
print(energy_types)
```

# Step 3: Insights:
### Linear Trends Across Decades
- Monthly trends across sectors:
    - Each monthly line chart shows a notable range of values per year, some more than others such as in the case of the Industrial Sector for Hydroelectric Power and Wood Energy between 1990 and 2000 and Hydroelectric Power in the Electric Power Sector from 1980 to 2020.
- Yearly trends across sectors:
    - In general most renewable energy consumption increased over time- the most recent trajectories tend to dip because full data for 2023 has not been collected.
- Linear Industrial Sector Observations:
    - Hydroelectric power and Biomas Losses experiences significant growth while other industries experienced stable energy consumption. The exception is Wood Energy which experienced a decline in the periond observed.
- Linear Residential Sector Observations:
    - Wood energy consumption experienced decline throughout the period studied with the exception of swift growth from about 1975 to 1985. Solar Energy increased, more rapidly in recent years, while Geothermal Energy remained stable.
- Linear Commercial Sector Observations:
    - Biomass Energy, Wood Energy, Geothermal Energy, Solar Energy, and Fuel Ethanol had notable growth over the years in the Commercial Sector. In more recent years, Waste Energy consumption looks to be in decline.
- Linear Transportation Sector Observations:
    - Every renewable energy type in the Transportation Sector experienced notable growth over time with the exception of "Other Biofuels" - hinting that the Transportation Sector has overall increasingly been utilizing more types of renewable energy.
- Linear Electric Power Sector Observations:
    - Hydroelectric Power experienced significant growth from 1950 until about the late 1970s where consumption entered a cycle of decrease and increase - not making significant growth in consumption since the late 1970s. Alternatively Wind Energy and Solar energy consumption has tremendous growth in the Electric Power Sector.
### Segmentation Trends Across Recent Months
- The Residential, Commercial, and Electric Power Sectors seem to experience cyclical changes in its renewable energy consumption. Unsurprisingly, Solar energy consumption decreases in the Fall/Winter months.
- While not cyclical, the segmentation of Renewable Diesel Fuel has slowly increased in the Transportation Sector.
- Hydroelectric Power seems cyclical for the Electric Power Sector as well - leading to a larger portion of the monthly energy consumed in the Winter Months.

### Segmentation Trends Over a Decade: 2010 v 2020
- The Industrial Sector experiences about a 20% decline in Wood Energy consumption and about a 20% increase in Solar Energy in this decade.
- The Transportation Sector saw a decrease in Fuel Ethanol by about 11% and saw an increase in Biodiesel and Renewable Diesel Fuel.
- In the Electric Power Sector we saw a notable increase in Solar Energy and Wind Energy while other sources of renewable energy experienced a decrease in the overall total percentage of consumption.
